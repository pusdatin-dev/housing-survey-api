**PROVINSI**

```python
# Endpoint API kamu
API_URL = "  https://59db4240a734.ngrok-free.app/api/v1/province"  # Ganti sesuai endpoint-mu

# Path file CSV di Google Drive ( BISA DIUBAH DARI PROVINSI - KELURAHAN )
CSV_FILE = "/content/drive/MyDrive/ProjectAPI/provinsi.csv"
```

```python
async def post_row(session, row):
    payload = {
        "id": int(row["code"]),
        "name": row["name"]
    }
    try:
        async with session.post(API_URL, json=payload, headers=headers) as response:
            resp_text = await response.text()
            print(f"POST {payload} => Status: {response.status} | Resp: {resp_text}")
    except Exception as e:
        print(f"Error posting {payload}: {e}")
```

```python
async def main():
    rows = []
    with open(CSV_FILE, newline='', encoding='utf-8') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            rows.append(row)

    semaphore = asyncio.Semaphore(20)  # Batas paralel request

    async with aiohttp.ClientSession() as session:
        tasks = []
        for row in rows:
            async def limited_post(row=row):
                async with semaphore:
                    await post_row(session, row)
            tasks.append(asyncio.ensure_future(limited_post()))
        await asyncio.gather(*tasks)

await main()
print("SELESAI mengirim semua data.")
```
